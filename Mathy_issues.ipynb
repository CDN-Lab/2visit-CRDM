```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
p = 0.3
beta = 0.22
A = 0.24
alpha = 0.435453

# Create an array of v values
v = np.linspace(-10, 10, 100)  # Adjust the range according to your needs

# Calculate sv values using the equation
sv = (p - beta * A/2) * v**alpha

# Plot sv as a function of v
plt.plot(v, sv, label='sv = (p - beta * A/2) * v^alpha')
plt.xlabel('v')
plt.ylabel('sv')
plt.title('Plot of sv as a Function of v')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define parameters
p = 0.3
beta = 0.22
A = 0.24
alpha = 0.445465


# Create an array of v values
svpos = np.linspace(0, 10, 100)  # Adjust the range according to your needs
svneg  = np.linspace(-10,0,100)

# Calculate sv values using the equation
vpos = (svpos / (p-beta*A/2)) ** (1 / alpha)
vneg = - (abs(svneg)/(p - beta*A/2)) ** (1/alpha)

# Plot sv as a function of v
plt.plot(svpos,vpos, svneg, vneg, label='v = (sv / (p-beta*A/2)) ** (1 / alpha)')
plt.xlabel('sv')
plt.ylabel('v')
plt.title('Plot of v as a Function of sv')
plt.legend()
plt.grid(True)
plt.show()
```
